```python
# Establece la conexión con la base de datos PostgreSQL
connection = psycopg2.connect(
    host="*******",
    port=5432,
    database="******",
    user="postgres",
    password="******"
)

# Consulta SQL
consulta_sql = '''SELECT student_id, grade, year, term, registration
FROM student_course
WHERE course_taken = 'BAIN081-14' AND grade > 1.0 AND year >= 2017 
ORDER BY year, term;
'''

df = pd.read_sql_query(consulta_sql, connection)

# 1279 ->  1142 (11%)

# Encontrar los student_id que cumplen con la condición df['registration'] != 'CURSADA'
student_id_cumplen_condicion = df[df['registration'] != 'CURSADA']['student_id']

# Eliminar las filas que tienen esos student_id
df = df[~df['student_id'].isin(student_id_cumplen_condicion)]
df = df.drop_duplicates(subset=['student_id'], keep='first')
df.head()
```

# Datos del arbol de pre-requisitos

```python
asignaturas = ['BAIN065-14', 'BAIN067-14', 'BAIN073-14', 'BAIN075-14']
resultados = []
for i in range(len(df)):
    datos = []
    datos.append(df.iloc[i,0])
    if(df.iloc[i,2] == 2020 or df.iloc[i,2] == 2021):
        datos.append(1)
    else:
        datos.append(0)
    # Obtener todo el registro académico anterior
    consulta_sql = "SELECT course_taken, grade, year, term, registration FROM student_course WHERE student_id = '" 
                    + str(df.iloc[i,0]) + "' AND (year < " + str(df.iloc[i,2]) + ' OR (year = ' + str(df.iloc[i,2]) 
                    + ' AND term < ' +  str(df.iloc[i,3]) + ")) ORDER BY year, term;"
    cursos = pd.read_sql_query(consulta_sql, connection);
    
    try:
        # Eliminar aquellas que tienen asignaturas convalidadas o reconocidas
        if(len(cursos[cursos['registration'] == 'RECONOCIDA']) == 0 and 
           len(cursos[cursos['registration'] == 'CONVALIDADA']) == 0):

            # Obtener todo el registro de promedios anteriores
            consulta_sql = "SELECT t_gpa, c_gpa, year, term FROM student_term WHERE student_id = '" + str(df.iloc[i,0]) 
                        + "' AND (year < " + str(df.iloc[i,2]) + ' OR (year = ' + str(df.iloc[i,2]) + ' AND term < ' 
                        +  str(df.iloc[i,3]) + ")) ORDER BY year, term, start_year;"
            promedios = pd.read_sql_query(consulta_sql, connection);
            psp1 = float('nan')
            psp2 = float('nan')
            # Calculando promedio de 2 semestres anteriores
            for year in range(df.iloc[i,2], 2013, -1):
                for term in [2,1]:
                    if(not math.isnan(psp1) and not math.isnan(psp2)):
                        break
                    if math.isnan(psp1):
                        try:
                            psp1 = promedios.loc[(promedios['year'] == year) & (promedios['term'] == term) 
                                                 & (promedios['t_gpa'] >= 1)].iloc[0,0]
                            PGA = promedios.loc[(promedios['year'] == year) & (promedios['term'] == term) 
                                                & (promedios['c_gpa'] >= 1)].iloc[0,1]
                            aux_year1, aux_term1 = year, term
                        except:
                            psp1 = float('nan')

                    else:
                        try:
                            psp2 = promedios.loc[(promedios['year'] == year) & (promedios['term'] == term) 
                                                 & (promedios['t_gpa'] >= 1)].iloc[0,0]
                            aux_year2, aux_term2 = year, term
                        except:
                            psp2 = float('nan')
            datos.append(psp1)
            datos.append(psp2)
            datos.append(PGA)
            
            # Si congeló en alguno de los 2 semestres anteriores
            congelado = 0
            if(df.iloc[i,3] == 1):
                if(aux_year1 == df.iloc[i,2] - 1 and aux_term1 == 2 and aux_year2 == df.iloc[i,2] - 1 and aux_term2 == 1):
                    pass
                else:
                    congelado = 1
            else:
                if(aux_year1 == df.iloc[i,2] and aux_term1 == 1 and aux_year2 == df.iloc[i,2] - 1 and aux_term2 == 2):
                    pass
                else:
                    congelado = 1
            datos.append(congelado)
            
            # información por asignatura
            for asignatura in asignaturas:
                filtro = (cursos['course_taken'] == asignatura)
                # Veces que se cursó la asignatura
                datos.append(filtro.sum())
                # última nota efectiva de la asignatura
                datos.append(cursos.loc[filtro & (cursos['grade'] > 1)]['grade'].values[-1])

                # Saber si fue cursado en pandemia
                aux = cursos[filtro]['year'].values[-1]
                if(aux == 2020 or aux == 2021):
                    datos.append(1)
                else:
                    datos.append(0)
                    
            #datos de ingreso
            consulta_sql = "SELECT program_id, start_year FROM student_program WHERE student_id = '" + str(df.iloc[i,0]) 
                            + "' AND start_year <"+ str(df.iloc[i,2]) +"ORDER BY start_year"
            demograficos = pd.read_sql_query(consulta_sql, connection)
            datos.append(demograficos.iloc[-1,0])
            datos.append(demograficos.iloc[-1,1])

            resultados.append(datos)
    except:
        pass
    
len(resultados)
```
